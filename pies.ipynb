```python
import pdal

def get_pipeline(vuelo, zona):
    return f'''{{
      "pipeline":[
        {{
          "type":"readers.pgpointcloud",
          "connection":"host='172.17.0.2' dbname='pointclouds' user='postgres' password='ventanuco' port='5432'",
          "table":"{vuelo}_zona{zona}"
        }},
        {{
            "type":"filters.sort",
            "dimension":"HeightAboveGround",
            "order":"DESC"
        }},
        {{
            "type":"filters.litree"
        }},
        {{
            "type":"writers.las",
            "filename":"data/PIES/pies_{vuelo}_zona{zona}.laz",
            "extra_dims":"all"
        }}
      ]
    }}'''
```

```python
p = pdal.Pipeline(get_pipeline('lida2', 2))
p.execute()
```
